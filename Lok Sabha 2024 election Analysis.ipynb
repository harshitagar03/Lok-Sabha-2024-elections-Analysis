# Exploratory Data Analysis

```python
order = All_State['Leading Party'].value_counts().index
plt.figure(figsize=(15, 10))
ax = sns.countplot(x='Leading Party', data=All_State,order = order, palette='viridis')

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Add labels and title
plt.xlabel('Leading Party')
plt.ylabel('Count')
plt.title('Count of Leading Party')

# Show plot
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Insight 1

#### Bharatiya Janata Party clearly shows the dominance followed by the Indian National Congress and Samajwadi Party.

```python
party_counts = All_State['Leading Party'].value_counts()

# Calculate percentage of each party
party_percentages = party_counts / party_counts.sum() * 100

# Group parties with less than 3% into 'Others'
threshold = 3  # Threshold percentage
mask = party_percentages < threshold
party_counts_combined = party_counts[~mask]
party_counts_combined['Others'] = party_counts[mask].sum()

# Plotting a pie chart
plt.figure(figsize=(10, 8))
plt.pie(party_counts_combined, labels=party_counts_combined.index, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

# Add labels and title
plt.title('Percentage Distribution of Leading Parties (Others < 3%)')

# Show plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
highest_margin_row = All_State.loc[All_State['Margin'].idxmax()]

# Get the corresponding candidate
candidate = highest_margin_row['Leading Candidate']
party = highest_margin_row['Leading Party']
cons = highest_margin_row['Constituency']
sta = highest_margin_row['State']
print("The candidate with the highest margin is:", candidate, "from", party, "Constituency:", cons, "State:", sta)
```
